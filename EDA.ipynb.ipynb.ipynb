## We can Observe that
### we can see here that lunch quality is effecting the all three scores of students

```python
#lets make 2 columsn with average and total scores of students
dff['total_score']= dff['math score']+dff['writing score']+dff['reading score']
dff['avg_score']=dff['total_score']/3
dff
```

```python
#lets draw a curve in which we are comparing if lunch is affecting female and male both the students
plt.subplots(1,2,figsize=(25,15))
plt.subplot(121)
sns.histplot(x=dff['avg_score'],bins=30,kde=True,color="aqua")
plt.subplot(122)
sns.histplot(x=dff['avg_score'], bins=40,kde=True,hue=dff["gender"])
```

## Conclusion --
### So we cna conclude that female have average marks better than males and their marks are also higher compare to males

```python
#lets draw this graph alag alag for females and males
plt.subplots(1,3,figsize=(20,10))
plt.subplot(131)
sns.histplot(data=dff[dff.gender=='female'] ,x='avg_score',bins=30, kde=True, hue="lunch")
plt.subplot(132)
sns.histplot(data=dff, x='avg_score',kde=True,bins=30,hue="lunch")
plt.subplot(133)
sns.histplot(data=dff[dff.gender=='male'],x='avg_score',bins=30,kde=True,hue="lunch")
```

## CONCLUSION
### We can conclude that male female both having standard lunch are performing better in their exams
### Their marks are higher and they have better average marks

```python
plt.subplots(1,3,figsize=(20,10))
plt.subplot(131)
sns.histplot(data=dff[dff.gender=='female'] ,x='avg_score',bins=30, kde=True, hue="parental level of education")
plt.subplot(132)
sns.histplot(data=dff, x='avg_score',kde=True,bins=30,hue="parental level of education")
plt.subplot(133)
sns.histplot(data=dff[dff.gender=='male'],x='avg_score',bins=30,kde=True,hue="parental level of education")
```
